```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(23)
```

```python
df = pd.DataFrame({'A': np.linspace(1, 10, 10)});df
```

```python
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))], axis=1); df
```

```python
df.iloc[0, 2] = np.nan
```

```python
def color_maiorque(s, coluna):
    maior = pd.Series(data=False, index=s.index)
    maior[coluna] = s.loc[coluna] >= 1
    return ['background-color: red' if maior.any() else '' for v in maior]
```

```python
def color_maiorque1(s):
    if s.B > 1.0:
        return ['background-color: yellow'] * 5
    else:
        return ['background-color: gray'] * 5
```

```python
df.style.apply(color_maiorque1, axis=1)
```

```python
# Abaixo de zero
def cor_negativo_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' %color
```

```python
df.style.applymap(cor_negativo_red)
```

```python
# Valores ou células especificos
def cor_max(s):
    maior = s == s.max()
    return ['background-color: red' if v else '' for v in maior]
```

```python
df.style.apply(cor_max)
```

```python
# Valores nulos - função Built - in
df.style.highlight_null(null_color='purple')
```

```python
# Mapa de calor
import seaborn as sns
cm = sns.light_palette("red", as_cmap=True)
```

```python
df.style.background_gradient(cmap='viridis')
```

```python
# Set properties
df.style.set_properties(**{'background-color': 'black',
                            'color': 'lawngreen',
                            'border-color': 'white'})
```

```python
# Grafico de barras
df.style.bar(subset=['A', 'B'], color='#d65f5f')
```

```python
# Estilo de tabela
from IPython.display import HTML

def hover(hover_color="#ffff99"):
    return dict(selector="tr:hover",
               props=[("background-color", "%s" % hover_color)])
styles = [
    hover(),
    dict(selector="th", props=[("font-size", "150%"),
    ("text-align", "center")])
]
html = (df.style.set_table_styles(styles)
       .set_caption("Hover to lighlight"))
html
```
